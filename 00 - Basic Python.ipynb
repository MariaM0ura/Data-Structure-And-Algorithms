```python
lista = []
for _ in range(1, 6):
  valor = int(input(': '))
  lista.append(valor)
```

```python
lista
```

```python
soma = 0
for i in range(0, 4):
  soma += lista[i]
```

```python
soma
```

```python
dic = {}
for _ in range(1, 4):
  nome = str(input(": "))
  nota = int(input(': '))
  dic[nome] = nota
```

```python
dic
```

```python
soma = 0
for notas in dic.values():
  soma += notas
soma
```
