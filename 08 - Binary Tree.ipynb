##Arvores Binárias
*   Combinação de vantagens de vetor ordenado e uma lista encadeada (busca rápida, inserção e eliminação)
*   os nós(cículos/valor) são conectadores por arestas(linhas) e tem no máximo dois filhos
*   Termologia de arvóres binárias:
    - caminho: ligação de um nó pra outro nó
    - Raiz: parte superior do nó, única e somente um caminho para outro nó
    - Pai: qualquer nó exeto a raiz, nós acima
    - Filho: nós abaixos
    - Folha: nó sem filhos
    - Subárvore: qualquer nó considerado como raiz de uma subárvore
    
* *Árvore de Busca*: a esquerda do nó o valor é menor e a direita do nó o valor é maior ou igual ao seu pai

###Árvores Binárias de Busca

```python
class NO:
    def __init__(self, valor):
        self.valor = valor
        self.esquerda = None                     ##endereços dos nós
        self.direita = None

    def mostra_no(self):
        print(self.valor)
```

```python
from re import S
class ArvoreBinariaDeBusca:
    def __init__(self):
        self.raiz = None
        self.ligacoes = []

    def inserir(self, valor):
        novo =  NO(valor)
        if self.raiz == None:
            self.raiz = novo
        else:
          atual = self.raiz                       ##endereço
          while True:
              pai = atual                         ##endereço
              ##esquerda
              if valor < atual.valor:
                  atual = atual.esquerda          ##atualização de Nó/Aresta
                  if atual == None:
                      pai.esquerda = novo
                      self.ligacoes.append(str(pai.valor)  + "->" + str(novo.valor))          ##vizualização da arvore por webgarfic
                      return

              else:
                  atual = atual.direita
                  if atual == None:
                      pai.direita = novo         ##atualização de Nó/Aresta
                      self.ligacoes.append(str(pai.valor)  + "->" + str(novo.valor))
                      return

    def pesquisar(self, valor):
        atual = self.raiz
        while valor != atual.valor:
            if valor < atual.valor:
                atual = atual.esquerda
            else:
                atual = atual.direita

            if atual == None:
                return None
        return atual

    #Raiz, Esquerda, Direita
    def pre_ordem(self, no):
        if no != None:
            print(no.valor)
            self.pre_ordem(no.esquerda)
            self.pre_ordem(no.direita)

    #Esquerda, Raiz, Direita
    def em_ordem(self, no):
        if no != None:
            self.em_ordem(no.esquerda)
            print(no.valor)
            self.em_ordem(no.direita)

    #Esquerda, direita, raiz
    def pos_ordem(self, no):
        if no != None:
            self.pos_ordem(no.esquerda)
            self.pos_ordem(no.direita)
            print(no.valor)

    def exclusao(self, valor):
        if self.raiz == None:
            print("Arvore vazia")
            return None

        atual =  self.raiz
        pai = self.raiz
        e_esquerda = True

        # pesquisa do valor
        while atual.valor != valor:
            pai = atual
            if valor < atual.valor:
                e_esquerda = True
                atual = atual.esquerda
            else:
                e_esquerda = False
                atual = atual.direita
            if atual == None:
                return None

        #o NO é uma folha
        if atual.esquerda == None and atual.direita == None:
            if atual == self.raiz:
                self.raiz = None
            elif e_esquerda == True:
                self.ligacoes.remove(str(pai.valor) + "->" + str(atual.valor))
                pai.esquerda == None
            else:
                self.ligacoes.remove(str(pai.valor) + "->" + str(atual.valor))
                pai.direita == None

        # o no a ser apagado não tem filho a esquerda
        elif atual.direita == None:
          if atual == self.raiz:
              self.raiz = atual.esquerda
          elif e_esquerda == True:
              pai.esquerda = atual.esquerda
          else:
              pai.direita = atual.esquerda

        #o no a ser apagado não tem filho a direita
        elif atual.esquerda == None:
            if atual == self.raiz:
                self.raiz = atual.direita
            elif e_esquerda == False:
                pai.esquerda = atual.direita
            else:
                pai.direita = atual.direita


        #o no possui dois filhos
        else:
            sucessor  = self.get_sucessor(atual)

            if atual == self.raiz:
                self.raiz = sucessor
            elif e_esquerda == True:
              pai.esquerda = sucessor
            else:
                pai.direita = sucessor





    def get_sucessor(self, no):
        pai_sucessor = no
        sucessor = no
        atual = no.direita
        while atual != None:
          pai_sucessor = sucessor
          sucessor = atual
          atual = atual.esquerda
        if sucessor != no.direita:
          pai_sucessor.esquerda = sucessor.direita
          sucessor.direita = no.direita
        return sucessor
```
